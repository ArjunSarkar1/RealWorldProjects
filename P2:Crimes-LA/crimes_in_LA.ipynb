```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

## Hours of Most Frequent Crimes

Convert times into int and extract only the hours.

```python
get_time = crimes["TIME OCC"].astype(int)
get_hour = get_time // 100
```

Get the counts for each hour.

```python
peak_crime_hour = get_hour.value_counts().sort_values(ascending = False)
```

Get the first index (hour with the most crimes).

```python
# print("The highest frequency of crimes occur at " + str(peak_crime_hour.index[0]) +
      # " (24-Hour-Format)")
```

## Most Frequent Night Crimes Location

Extracting rows according to the night time period


```python
night_times = crimes[(get_hour >= 22) | (get_hour < 4)]
```

Extracting all locations during the night time

```python
get_locations = night_times["AREA NAME"]
```

Getting the frequency of each location in descending order

```python
freq_location = get_locations.value_counts().sort_values(ascending = False)
```

Getting our peak night crime location.

```python
peak_night_crime_location = freq_location.index[0]
print("The peak night location is " + peak_night_crime_location)
```

## Crimes Among Different Age Groups

Storing the `Vict Age` column in a variable.

```python
all_victim_ages = crimes["Vict Age"]
```

Defining bins for age groups.

```python
crime_age_bins = [0,17,25,34,44,54,64, np.inf]
```

Labeling the age groups.

```python
crime_age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
```

Adding another column in `crimes` dataset called `Age Groups`.

```python
crimes["Age Groups"] = pd.cut(all_victim_ages,
                              bins=crime_age_bins,
                              labels=crime_age_labels)
```

Get the frequency of each age group.

```python
victim_ages = crimes["Age Groups"].value_counts().sort_index()
victim_ages
```
